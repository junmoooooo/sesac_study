```python
# 4칙연사 계산기 5대 운영(1,2,3,4,5)

# 메뉴
# 1.계산하기
# 2.각 계산기의 마지막 결과 확인

# 1.계산하기

# 계산기선택:(1,2,3,4,5 중에 택일)

# num1:
# num2:
# 기호: 
# (정수(음수 포함)와 실수외에는 입력이 되지않도록 설정 - 일반 함수로 정의할 것)
# 나누기 같은 경우에는 자식 클래스를 만들어서 분모가 0일 때 결과값이 0이 되도록 메소드 오버라이드 할 것

# 2. 각 계산기의 마지막 결과 확인

# 계산기선택:(1,2,3,4,5 중에 택일)



def num_input(i):
    while True:
        num=input("num{} : ".format(i))
        a=0
        for c in num:
            if c not in "0123456789-.":
                print("잘못입력하셨습니다. 다시 입력해주세요.")
                a=1
                break
        if a==0:
                       
            if num.count('-') > 1  or num.count('.') > 1:
                print("잘못입력하셨습니다. 다시 입력해주세요.")
                continue
            if num.count('-') == 1:
                if num[0] != '-':
                    print("잘못입력하셨습니다. 다시 입력해주세요.")
                    continue
            if num.count('.')==1:
                return float(num)
            else:
                return int(num)
            
#정수와 실수만 입력 받도록 설정
                
class Fourcal:
    def __init__(self,a,b,c):
        self.a=a
        self.b=b
        self.c=c
        self.result=0
    def fourcal(self):
        if self.c == '+':
            self.result=self.a+self.b
        if self.c == '-':
            self.result=self.a-self.b
        if self.c == '*':
            self.result=self.a*self.b
        return self.result
    def div(self):
        self.result=self.a/self.b
        return self.result
class SubFourcal(Fourcal):
    def div(self):
        if self.b==0:
            self.result=0
        else:
            self.result=self.a/self.b
        return self.result
        
        
menu='''1.계산하기
2.각 계산기의 마지막 결과 확인
3.나가기
선택 : '''

while True:
    choice = input(menu)
    if choice == '3':
        print("종료합니다")
        break

    if choice == '1':
        while True:
            calnum=input("계산기를 선택하세요.(a,b,c,d,e) : ")
            if calnum not in "abcde":
                continue
            num1=num_input(1)
            num2=num_input(2)
            cal=input("입력할 연산 기호를 입력하시오")
            if calnum == 'a':
                a=SubFourcal(num1,num2,cal)
                if cal not in "+-*":
                    print(a.div())
                else:
                    print(a.fourcal())
            if calnum == 'b':
                b=SubFourcal(num1,num2,cal)
                if cal not in "+-*":
                    print(b.div())
                else:
                    print(b.fourcal())
            if calnum == 'c':
                c=SubFourcal(num1,num2,cal)
                if cal not in "+-*":
                    print(c.div())
                else:
                    print(c.fourcal())
            if calnum == 'd':
                d=SubFourcal(num1,num2,cal)
                if cal not in "+-*":
                    print(d.div())
                else:
                    print(d.fourcal())
            if calnum == 'e':
                e=SubFourcal(num1,num2,cal)
                if cal not in "+-*":
                    print(e.div())
                else:
                    print(e.fourcal())
            break
    if choice == '2':
            calnum=input("계산기를 선택하세요.(a,b,c,d,e) : ")
            if calnum == 'a':
                print(a.result)
            if calnum == 'b':
                print(b.result)
            if calnum == 'c':
                print(c.result)
            if calnum == 'd':
                print(d.result)
            if calnum == 'e':
                print(e.result) 
```
